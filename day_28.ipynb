## Stream plots visualize vector field data, representing the flow of a fluid or movement in a system, ideal for understanding fluid dynamics or electromagnetic fields. They're used when analyzing continuous directional data, offering insights into the direction and intensity of flow.

```python
# Generate grid of points
Y, X = np.mgrid[-3:3:100j, -3:3:100j]

# Define vector field
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Plot streamlines
plt.streamplot(X, Y, U, V, density=[0.5, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Stream Plot Example')
plt.show()
```
