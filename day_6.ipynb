```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 6 Visualizing "Dist plot"

```python
df = sns.load_dataset("tips")
df.head()
```

## Distplot is a versatile tool for visualizing the distribution of a single numerical variable, providing insights into its central tendency, spread, and skewness. It's especially useful when exploring the shape and characteristics of data distributions, aiding in data exploration, outlier detection, and hypothesis testing.

```python
# create distplot
sns.distplot(a=df["tip"],kde = False)
plt.title("Distribution of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Density")
```


# The KDE (Kernel Density Estimation) in a distplot represents an estimation of the probability density function of the underlying data distribution.

```python
sns.distplot(a=df["total_bill"], kde = True)
plt.title("Distribution of total_bill")
plt.xlabel("bill Amount")
plt.ylabel("Density")
```

```python
# random varibale distplot
a = np.random.rand(100)
sns.distplot(a=a)
plt.title("Distribution Plot of Random Data")
plt.xlabel("Values")
plt.ylabel("Density")
```
