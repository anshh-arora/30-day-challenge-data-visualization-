```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 23 visualizind "Clustered Bar Plot"

```python
df= sns.load_dataset("iris")
df.head()
```

## A clustered bar plot is used to compare multiple categories across different groups, displaying bars for each category side by side for each group. It is ideal for visualizing relationships and differences within and between groups. To create one, use pandas or matplotlib to plot data with groupby or pivot tables, ensuring each bar represents a distinct category within grouped data.

```python
# Aggregate the data to get mean sepal and petal lengths for each species
agg_df = df.groupby('species').agg({
    'sepal_length': 'mean',
    'petal_length': 'mean'
}).reset_index()

# Melt the DataFrame for easier plotting
plot_df = agg_df.melt(id_vars='species', var_name='Measurement', value_name='Mean Value')
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))
sns.barplot(data=plot_df, x='species', y='Mean Value', hue='Measurement')

# Customize the plot
plt.title('Mean Sepal and Petal Lengths by Species')
plt.xlabel('Species')
plt.ylabel('Mean Value')

# Show the plot
plt.show()
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))
ax= sns.barplot(data=plot_df, x='species', y='Mean Value', hue='Measurement')
for bars in ax.containers:
    ax.bar_label(bars)
# Customize the plot
plt.title('Mean Sepal and Petal Lengths by Species')
plt.xlabel('Species')
plt.ylabel('Mean Value')

# Show the plot
plt.show()
```
