```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 4 Visualizing "Scatter plot"

```python
# Load the "iris" dataset from Seaborn library
df = sns.load_dataset("iris")
df.head()  # Display the first few rows of the dataset for a quick look
```

## Scatter plots are used to visualize the relationship between two numerical variables. They display individual data points as dots on a two-dimensional plane, with one variable plotted along the x-axis and the other plotted along the y-axis.

```python
sns.scatterplot(data = df)
```

The above graph contain all the columns of iris dataset

```python
#now visualize species column which contain 3 type of variable
sns.scatterplot(data=df.species)
```

the above graph shows that "setosa" species are in the range of 0 to 60 , "versicolor" are in the range of 45 to 105 and "virginica" are in the range of 100 to 150

```python
## Now we use some random numbers to learn more about scatter plot
np.random.seed(0)
x= np.random.rand(100)
y = np.random.rand(100)
# Create a random categorical variable for hue
category = np.random.choice(['A', 'B'], size=100)
print (x)
print(y)
sns.scatterplot(x=x,y=y, hue=category)
```
