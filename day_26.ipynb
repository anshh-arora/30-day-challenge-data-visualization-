## Day 26 of Visualizing "Network Graph"

```python
df = sns.load_dataset("tips")
df.head()
```

## A network graph is a visualization that represents connections or relationships between entities as nodes and edges. It's used to analyze and visualize complex relationships, such as social networks, transportation systems, or biological networks. To create one, use Python libraries like NetworkX or Graph-tool, where nodes represent entities and edges represent connections, and visualize using matplotlib or other plotting libraries.

```python
# Create a graph
G = nx.from_pandas_edgelist(df, source='day', target='smoker')

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

This graph visualizes the connections between days and smoking status in the dataset, with nodes representing days and smokers/non-smokers, and edges indicating instances of these combinations.

```python
G = nx.from_pandas_edgelist(df, source="tip", target="time")

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```


This graph illustrates relationships between tip amounts and meal times, with nodes representing tip values and meal times, and edges indicating connections between these entities.
