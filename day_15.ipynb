```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## day 15 visualizing "pair grid"

## A PairGrid in Seaborn is a versatile tool for creating a grid of subplots where each cell contains a plot for the relationship between two variables. It allows for the visualization of pairwise relationships across multiple variables in a dataset.

```python

diamonds = sns.load_dataset('diamonds')
diamonds.head()


```

```python
# Create a PairGrid
grid = sns.PairGrid(data=diamonds)
# Populate the PairGrid with scatter plots
grid.map(plt.scatter)
```

```python
# lets use minimul data points to visualize easily
min = diamonds.head(50)
min.head()
```

```python
# Selecting specific variables
selected_variables = min[["carat", "price"]]

sns.PairGrid(data=selected_variables)
# Populate the PairGrid with scatter plots
grid.map(plt.scatter)
```
