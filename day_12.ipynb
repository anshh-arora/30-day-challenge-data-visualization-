```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 12 visualizing "Swarm Plot"

```python
df = sns.load_dataset("tips")
df.head()
```

## Use swarm plots when visualizing categorical data with many levels or overlapping points, enhancing readability by spreading points apart. Adjust point size and jitter for better visualization of dense data points.

```python
# Selecting the first 30 rows of the DataFrame
df_subset = df.head(30)

# Creating a swarm plot using Seaborn for the subset of data
sns.swarmplot(data=df_subset, x='total_bill', y='tip',hue="sex")
sns.set(rc={"figure.figsize":(19,4)})
```

you can see the dots with blue represent "Male" and other represent "Female". the graph shows the relationship between tip and totalbill.

```python
# Selecting the first 30 rows of the DataFrame
df_subset = df.head(30)

# Creating a swarm plot using Seaborn for the subset of data
sns.swarmplot(data=df_subset, x='tip', y='size',hue="sex")
sns.set(rc={"figure.figsize":(19,4)})
```

the above graph shows that the tip given by male and female.
