```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## day 13 visualizing "Stacked Barplot"

```python
df = sns.load_dataset("iris")
df.head()
```

## Stacked barplots are used to visualize multiple categories of data where each category is broken down into sub-categories. They display the total amount for each category while also showing the contribution of each sub-category to the total. Stacked barplots help in comparing proportions across different groups and identifying patterns or trends within the data.

```python
#creating stacked barplot
min_df = df.head(1000)
sns.barplot(data=min_df, x="sepal_width", y="petal_width", hue="species")

# Add labels and title
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('Stacked Bar Plot')

# Show the plot
```
