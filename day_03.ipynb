```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Day 3 Visualizing "Piechar" and "Donut Chart"

```python
# Load the "tips" dataset from Seaborn library
df = sns.load_dataset("tips")
df.head()  # Display the first few rows of the dataset for a quick look
```

## Pie charts are effective for displaying proportions within a whole, allowing quick comparisons of categorical data distributions.

```python
# pie chart of day columns
day_counts = df.day.value_counts()

label = day_counts.index #name 
size = day_counts.values #percentage value

plt.pie(x=size, labels=label, autopct='%1.1f%%')
plt.axis("equal")
plt.title("A pie chart")
plt.show()
```

```python
# pie chart of sex columns
sex_counts = df.sex.value_counts()

label = sex_counts.index #name 
size = sex_counts.values #percentage value

plt.pie(x=size, labels=label, autopct='%1.1f%%')
plt.axis("equal")
plt.title("A pie chart of sex distribution")
plt.show()
```

##  Donut charts, a variant of pie charts with a central hole, offer the same insights with additional space for annotations or labels, enhancing clarity and visual appeal.

```python
#donut chart of day column
day_counts = df.day.value_counts()

label = day_counts.index
size = day_counts.values

plt.pie(x=size,labels = label,autopct = "%1.1f%%", wedgeprops =dict(width = 0.3))
```

```python
#donut chart of day column
day_counts = df.day.value_counts()

label = day_counts.index
size = day_counts.values

plt.pie(x=size,labels = label,autopct = "%1.1f%%", radius=1.5)
plt.pie([1],colors=["w"],radius = 1)
```

```python
# pie chart of sex columns
sex_counts = df.sex.value_counts()

label = sex_counts.index #name 
size = sex_counts.values #percentage value

plt.pie(x=size, labels=label, autopct='%1.1f%%', radius=1.5)
plt.pie([1],colors=["w"],radius=1)
plt.axis("equal")
plt.title("A pie chart of sex distribution")
plt.show()
```
