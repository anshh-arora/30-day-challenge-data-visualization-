```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 17 Visualizing "Scatter plot Matrix"

```python
df = sns.load_dataset("iris")
df.head()
```

## A scatter plot matrix is like a grid of scatter plots, where each plot shows the relationship between two numerical variables. It helps us see how variables relate to each other and if there are any patterns or trends in the data.








1. Scatter plot matrix displays pairwise relationships between multiple numerical variables.
2. It helps visualize correlations, clusters, and outliers in the dataset.
3. Useful for exploratory data analysis to identify patterns and understand the data's underlying structure.


```python
#creating a normal scatter plot
sns.scatterplot(data=df,x=df.sepal_width,y=df.petal_width)
```

```python
# Create a scatter plot matrix
sns.pairplot(df)
```

```python
# Create a pair plot with hue for a categorical variable
sns.pairplot(df, hue='species', markers=['o', 's', 'D'])
```

## A scatter plot matrix is essentially the same as a pair plot (pairplot). Both terms refer to a grid of scatter plots where each plot shows the relationship between two numerical variables in a dataset.

```python
# Add regression lines to scatter plots
sns.pairplot(df, kind='reg', hue='species', markers=['o', 's', 'D'])
```

```python
# Add additional customization as needed, such as adjusting plot size, labels, and titles
sns.pairplot(df, height=3, aspect=1.5, hue="species")
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)

plt.show()
```
