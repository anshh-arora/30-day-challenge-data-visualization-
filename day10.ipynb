```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## day 10 visualizing 'Pair Plot'

```python
df = sns.load_dataset("tips")
df.head()
```

## Pairplot is used to visualize relationships between pairs of variables in a dataset, making it easy to spot patterns and correlations. It's particularly useful for exploratory data analysis to understand the overall structure and associations within the data.

```python
sns.pairplot(data=df)
```

The above pair plot shows the relationship with other numerical variables this can be very useful while working on any Mcahine Learning Porject

```python
sns.pairplot(df,hue="time")
```
