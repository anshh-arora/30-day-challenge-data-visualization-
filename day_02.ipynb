```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Day 2 Visualizing "Histplot" and "Countplot"

```python
# Load the "tips" dataset from Seaborn library
df = sns.load_dataset("tips")
df.head()  # Display the first few rows of the dataset for a quick look
```

## Histplot: Visualizes the distribution of a single numerical variable through a histogram, showing the frequency or density of data points within predefined bins.

```python
# create a histogram 
ax = sns.histplot(data=df["day"])
for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see that  most of the customers visit our restaurant on Saturday and Sunday.

## Countplot: Displays the count of observations in each category of a categorical variable, providing a quick overview of the distribution of categorical data.







```python
# count plot
ax = sns.countplot(data=df.sex)
for bars in ax.containers:
    ax.bar_label(bars)
```

The graph show us that most of the customers are Male as compair to Female.

```python
sns.histplot(data=df.smoker)
```

This shows that most of the person are non smoker.
