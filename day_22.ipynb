```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 5 Visualizing "violin plot with hue"

```python
# Load the "iris" dataset from Seaborn library
df = sns.load_dataset("iris")
df.head()  # Display the first few rows of the dataset for a quick look
```

## Violin plots display the distribution of data across different levels of a categorical variable, combining the features of box plots and kernel density plots. They are ideal for visualizing the distribution, central tendency, and variability of data, especially when comparing multiple groups or categories.

```python
# Creating a normal Swarm Plot
sns.violinplot(data=df)
plt.title("Violin PLot")
plt.ylabel("count")
plt.xlabel("name of the columns")
```

```python
# Create a violin plot with hue
sns.violinplot(x='species', y='sepal_length', hue='species', data=df, split=True)

# Set the title and labels for the plot
plt.title("Violin Plot")
plt.ylabel("Sepal Length")
plt.xlabel("Species")

# Show the plot
plt.show()
```
