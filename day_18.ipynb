```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 18 visualizing "Area Plot"

```python
df=sns.load_dataset("titanic")
df.head()
```


## Area plot visually represents the cumulative sum of numerical values over a continuous interval, typically time. It's used to showcase trends, proportions, and changes in data over time or across categories. Ideal for illustrating data evolution and comparing proportions in a dataset.

```python
# Create an area plot
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df.index, y="age", estimator="sum", ci="deprecated")
plt.title("Cumulative Age Distribution of Titanic Passengers")
plt.xlabel("Index")
plt.ylabel("Cumulative Age")
plt.show()
```

```python
# creating area plot for cumulative pclass distribution of titanic passenger
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df.index, y=df.groupby('pclass').cumcount()+1)
plt.title("Cumulative Pclass Distribution of Titanic Passengers")
plt.xlabel("Index")
plt.ylabel("Cumulative Count")
plt.show()
```

calculates the cumulative count of each passenger class using df.groupby('pclass').cumcount(), and then adds 1 to ensure that the count starts from 1 instead of 0. 

```python
# Create an area plot for fare 
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df.index, y="fare", estimator="sum", ci="deprecated")
plt.title("Cumulative fare Distribution of Titanic Passengers")
plt.xlabel("Index")
plt.ylabel("Cumulative fare")
plt.show()
```

```python
# lets take only 50 data point to visualizse clearly
new_data= df.head(50)
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=new_data, x=new_data.index, y="fare", estimator="sum", ci="deprecated", hue="sex", palette=["Red","black"])
plt.title("Cumulative fare Distribution of Titanic Passengers")
plt.xlabel("Index")
plt.ylabel("Cumulative fare")
plt.show()
```
