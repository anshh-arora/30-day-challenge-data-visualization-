```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from wordcloud import WordCloud
```

## Day 16 visualizing "Word cloud"

```python
df = sns.load_dataset("iris")
df.head()
```

## A word cloud is a visual representation of text data where words are displayed in varying sizes based on their frequency. It's used to quickly identify key terms and themes within a body of text, aiding in summarization and analysis.

```python
# Convert the species column to a single string
text = ' '.join(df['species'])
```

```python
# Generate the word cloud
wordcloud = WordCloud(width=700, height=300, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Larger words represent species that occur more frequently in the dataset, while smaller words represent species that occur less frequently.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#converting class into single string
text_1 = " ".join(titanic["class"])
```

```python
#creating word cloud
wordcloud_1 = WordCloud(width=800, height=400, background_color="white").generate(text_1)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_1,)
plt.axis("off")
plt.show()
```

```python
# Concatenate multiple columns into a single string
text_combined = titanic[['class', 'age', 'sex']].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

# Join the strings in the Series into a single string
text_combined = ' '.join(text_combined)

# Now, text_combined contains the concatenated values of 'class', 'age', and 'sex' columns
#creating word cloud
wordcloud_2 = WordCloud(width=800, height=400, background_color="white").generate(text_combined)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_2,)
plt.axis("off")
plt.show()
```

```python
#converting alone column from bollean to string
text_alone = " ".join(titanic["alone"].astype(str))
```

```python
#creating a word cloud
word_alone = WordCloud(width=900, height=700, background_color="#200589", prefer_horizontal=0.9, max_font_size=1000, random_state=42).generate(text_alone)

plt.figure(figsize=(10,5))
plt.imshow(word_alone)
plt.axis("off")
plt.show()
```
