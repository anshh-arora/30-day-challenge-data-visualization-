```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 21 visualizing "Swarm Plot with hue"

```python
df = sns.load_dataset("tips")
df.head()
```

## In a swarm plot, the hue parameter adds another categorical variable by using different colors to distinguish data points. This helps in comparing groups within the same category. You use hue by specifying it in the sns.swarmplot function, like this: sns.swarmplot(x='category', y='value', hue='group', data=df).









```python
# Selecting the first 30 rows of the DataFrame
df_subset = df.head(30)

# Creating a swarm plot using Seaborn for the subset of data
sns.swarmplot(data=df_subset, x='total_bill', y='tip',hue="sex")
sns.set(rc={"figure.figsize":(19,4)})
```

```python
# Selecting the first 30 rows of the DataFrame
df_subset = df.head(30)

# Creating a swarm plot using Seaborn for the subset of data
sns.swarmplot(data=df_subset, x='tip', y='size',hue="sex")
sns.set(rc={"figure.figsize":(19,4)})
```
