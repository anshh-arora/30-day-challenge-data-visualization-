```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## day 14 visualizing "Line Plot with Error Bars"

```python
df=sns.load_dataset("titanic")
df.head()
```

## Line plots with error bars are often used when we want to visualize the trend or relationship between two variables over a continuous range. The error bars provide a visual representation of the variability or uncertainty associated with each data point, making it easier to interpret the reliability of the trend.

```python
#creating Line Plot with Error Bars
sns.lineplot(data=df,x="survived",y="age",errorbar=('ci', 95))
```

This visualization aims to show how the average age varies across different survival statuses

```python
#creating Line Plot with Error Bars
sns.lineplot(data=df,x="fare",y="age",errorbar=('ci', 95),hue="sex")
```

line plot that visualizes the relationship between fare paid and age, with error bars indicating the variability in age at different fare levels. Additionally, it uses different colors to represent male and female individuals, allowing for a comparison of age trends between the two genders.

```python
#lets take 50 rows
less_data= df.head(50) 
sns.lineplot(data=less_data,x="fare",y="pclass",errorbar=('ci', 95),hue="sex")
```

 this plot provides a comprehensive visualization of the relationship between fare, passenger class, and gender, helping to identify patterns and trends within the data.
