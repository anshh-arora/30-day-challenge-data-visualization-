```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 7 Visualizing "jointplot(scatter and hist)"

```python
df = sns.load_dataset("tips")
df.head()
```


## Jointplot in Seaborn offers a convenient way to visualize the relationship between two numerical variables through scatter plots and histograms

```python
#creating joinplot
sns.jointplot(data=df, x="total_bill", y="tip",kind = "scatter")
```

##  It's particularly useful for exploring correlations, identifying patterns, and understanding the joint distribution of variables, aiding in data exploration and analysis tasks.








```python
# let use some random variables
x=np.random.rand(50)
y =np.random.rand(50)
category = np.random.choice(["A","B"], size=50)
sns.jointplot(x=x,y=y, hue = category, kind="scatter")
```

```python
#creating joinplot
sns.jointplot(data=df, x="tip", y="size",kind = "scatter",color = "red")
```
