```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# Tips
1. Use barplots when comparing categories or groups, such as showing the distribution of different classes or comparing quantities across discrete categories.2. 
Use line plots when visualizing trends over time or across ordered categories, ideal for displaying continuous data and highlighting changes or patterns.

```python
# Load the "iris" dataset from Seaborn library
df = sns.load_dataset("iris")
df.head()  # Display the first few rows of the dataset for a quick look
```

## Barplots are used to compare categories or groups, showing the distribution or proportion of data with rectangular bars.


```python
# Create a barplot using seaborn
ax= sns.barplot(data=df)
for bars in ax.containers:
    ax.bar_label(bars)
# the above for loop help to add number on the each bars. it make visualization easier
```

## Line plots, on the other hand, are ideal for visualizing trends over time or across ordered categories, connecting data points with lines to highlight changes.

```python
# line plot of complete data set
sns.lineplot(data=df)
```

```python
# Line plot of single column ("sepal_length")
sns.lineplot(data=df.sepal_length)
```
