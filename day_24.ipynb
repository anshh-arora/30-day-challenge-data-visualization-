```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

## Day 24 Visualizing "Heatmap with Annotations"

```python
df = sns.load_dataset("titanic")
df.head()
```

## A heatmap with annotations is used to visualize data where individual values are represented as colors in a matrix, with annotations displaying the actual data values on the cells. It is ideal for spotting patterns, correlations, and anomalies within a dataset at a glance.

```python
# Create a pivot table for survival rate by class and gender
pivot_table = df.pivot_table(index='class', columns='sex', values='survived', aggfunc='mean')
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu")

# Customize the plot
plt.title('Survival Rate by Class and Gender')
plt.xlabel('Gender')
plt.ylabel('Class')

# Show the plot
plt.show()
```

```python
# Create a pivot table for passenger count by class and embarkation port
pivot_table = df.pivot_table(index='class', columns='embark_town', values='survived', aggfunc='count')
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")

# Customize the plot
plt.title('Passenger Count by Class and Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Class')

# Show the plot
plt.show()
```
